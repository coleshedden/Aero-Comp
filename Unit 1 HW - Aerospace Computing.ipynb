```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Custom Function to input into contour function
    
def cust_funct(A, B, n, X, m, Y, phi, psi):
    
    # Equation for F
    
    F = A*np.cos(n*np.pi*X + phi) + B*np.cos(m*np.pi*Y + psi)
    return F
```

```python
# Function to make a contour plot

def make_a_contour(x0, x1, y0, y1):

    # Create 1-D arrays for x and y
    
    x = np.linspace(x0, x1, 50)
    y = np.linspace(y0, y1, 50)
    
    # Create array for F 
    
    F = np.zeros((len(y), len(x)))
    
    # Input values of F for corresponding x and y coordinates
    
    for i in range(len(y)):
        for k in range(len(x)):
            F[i][k] =  cust_funct(A, B, n, x[k], m, y[i], phi, psi)
            
    # Create figure and axis objects
    
    fig, ax = plt.subplots()
    
    # Plot x, y and F with viridis color theme
    
    contour = ax.contourf(x, y, F, cmap="viridis")
    
    # Add colorbar, title, axes, labels
    
    fig.colorbar(contour)
    ax.set_title("Contour Plot of F(x, y) = A*cos(n*pi*x + phi) + B*cos(m*pi*y + psi)")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    
    
    # Adjust window of graph based on user input
    ax.axis([x0, x1, y0, y1])
    
    return fig, ax
```

```python
# Lineplot function

def make_a_lineplot(x, F):
    
    # Create figure and axis objects
    
    fig, ax = plt.subplots()
    
    # Plot graph of F
    
    line = plt.plot(x, F)
    
    # Set plot title
    
    ax.set_title("Line Plot of F(x) = A*cos(n*pi*x + phi) + B*cos(m*pi*y + psi)")

    
    return fig, ax
    
```
